##Graphs max/min
Elliptic Parabaloid's opening in both directions with min and max at points (0, 0).

```
def f(x, y): 
  return (x ** 2 + y ** 2) 
  
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.plot([0], [0], [0], markerfacecolor='r', markeredgecolor='r', marker='o', markersize=15, alpha=1)
ax.set_title('Surface');
ax.view_init(20, 35)
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
